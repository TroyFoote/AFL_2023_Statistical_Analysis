Q2. Has the style of game play reduced or increased the number of goals being scored?

```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Data Files
season_summary_path = "Resources/Season_Summary.csv"
```

```python
# Read in the Season_Summary file
season_summary_complete_df = pd.read_csv(season_summary_path)
season_summary_complete_df.head()
```

```python
# Strip the spaces after the column names
season_summary_complete_df.columns = [x.strip() for x in season_summary_complete_df.columns]
season_summary_complete_df.head()
```

```python
#check datatypes of each column
print(season_summary_complete_df.dtypes)
```

```python
# Reame column headers
season_summary_renamed_col = season_summary_complete_df.rename(columns={"Year":"Season", "Premier_LP":"Premier_Ladder_Position", "Runnerup":"R_Up",
                                                                        "Runnerup_LP":"R_Up_Ladder_Position", "Last":"Wooden_Spoon", "Unnamed: 8":"GK_Team", 
                                                                        "Unnamed: 9":"Goals_Kicked", "Unnamed: 11":"D_Team", "Unnamed: 12":"Disposals"})
season_summary_renamed_col.head()
```

```python
# Create DataFrame of just the GoalKickers columns
goal_kicker_df = season_summary_renamed_col[["Season","Most Goals","GK_Team","Goals_Kicked"]]
goal_kicker_df.head()
```

```python
# Create DF of just the Disposals columns
disposals_df = season_summary_renamed_col[["Season","Most Disposals","D_Team","Disposals"]]
disposals_df.head()
```

```python
#Save csv files
goal_kicker_df.to_csv('created_files/goal_kicker_data.csv', index=False)
disposals_df.to_csv('created_files/disposals_data.csv', index=False)
```
