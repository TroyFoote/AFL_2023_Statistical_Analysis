```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Data Files
scoring_summary_path = "Resources/Scoring_Summary.csv"
```

```python
# Read in the Season_Stats file
scoring_summary_complete_df = pd.read_csv(scoring_summary_path)
scoring_summary_complete_df.head()
```

```python
#check datatypes of each column
print(scoring_summary_complete_df.dtypes)
```

```python
# Strip the spaces after the column names
scoring_summary_complete_df.columns = [x.strip() for x in scoring_summary_complete_df.columns]
scoring_summary_complete_df.head()
```

```python
# Reame column headers
scoring_summary_renamed_col = scoring_summary_complete_df.rename(columns={"Year":"Season", "GM":"Games_Played", "G":"Goals", "B":"Behinds", "Tot":"Scoring_Shots", "Ave.":"Scoring_Shots_per_Game", "GK%":"Goal_Kicking_Accuracy"})
scoring_summary_renamed_col.head()
```

```python
# Create DF of just the required columns
goals_df = scoring_summary_renamed_col[["Season","Games_Played","Goals"]]
goals_df.head()
```

```python
goals_df["Ave_goals"] = goals_df.Goals / goals_df.Games_Played
```

```python
goals_df.head()
```

```python
# Bring in the goal kicker csv File
goal_kicker_summary_path = "csv_files/goal_kicker_data.csv"

# Read in the Season_Stats file
goal_kickers_df = pd.read_csv(goal_kicker_summary_path)
goal_kickers_df.head()
```

```python
# Merge the two dataframes
goals_merged_df = pd.merge(goals_df, goal_kickers_df, on='Season')
goals_merged_df.head()
```

```python
goals_merged_df["%_of_goals"] = (goals_merged_df.Goals_Kicked / goals_merged_df.Goals) * 100
goals_merged_df.head()
```
