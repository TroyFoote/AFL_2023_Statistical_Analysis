Q1. Has the game become a higher or lower disposal(faster or slower moving) game since the beginning of the AFL Competition in 1990?

```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Data Files
season_stats_path = "Resources/Season_Stats.csv"
```

```python
# Read in the Season_Stats file
season_stats_complete_df = pd.read_csv(season_stats_path)
season_stats_complete_df.head()
```

```python
#check datatypes of each column
print(season_stats_complete_df.dtypes)
```

```python
# Strip the spaces after the column names
season_stats_complete_df.columns = [x.strip() for x in season_stats_complete_df.columns]
season_stats_complete_df.head()
```

```python
# Reame column headers
season_stats_renamed_col = season_stats_complete_df.rename(columns={"YEAR":"Season", "KI":"Kicks", "MK":"Marks", "HB":"Handballs", "DI":"Total_Disposals",
                                                "GL":"Goals", "BH":"Behinds", "HO":"Hit_Outs", "TK":"Tackles", "FF":"Free For", "BR":"Brownlow_Votes", "TEAMS":"Teams"})
season_stats_renamed_col.head()
```

```python
# Create DF of just the required columns
disposals_df = season_stats_renamed_col[["Season","Total_Disposals","Teams"]]
disposals_df.head()
```

```python
disposals_df["Ave_disposals"] = disposals_df.Total_Disposals / disposals_df.Teams
```

```python
disposals_df.head()
```

```python
# Bring in the disposals csv File
disposals_summary_path = "created_files/disposals_data.csv"

# Read in the Season_Stats file
player_disposals_df = pd.read_csv(disposals_summary_path)
player_disposals_df.head()
```

```python
# Merge the two dataframes
disposals_merged_df = pd.merge(disposals_df, player_disposals_df, on='Season')
disposals_merged_df.head()
```

```python
# Find percentage of disposals by leading disposal getter per year
disposals_merged_df["%_of_disposals"] = (disposals_merged_df.Disposals / disposals_merged_df.Total_Disposals) * 100
disposals_merged_df
```

```python
#Save csv files
disposals_merged_df.to_csv('created_files/Disposals_DataFrame.csv', index=False)
```
